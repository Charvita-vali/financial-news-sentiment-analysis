```python
import sys
import os

sys.path.append(os.path.abspath(".."))

from src.data_preprocessing import load_and_prepare_data

df = load_and_prepare_data("../data/real_financial_news.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

sentiment_counts = df['sentiment_label'].value_counts()

plt.figure(figsize=(8,5))
bars = plt.bar(sentiment_counts.index, sentiment_counts.values)

plt.title("Distribution of Financial News Sentiment (Real-World Dataset)", fontsize=16, fontweight='bold')
plt.xlabel("Sentiment Label")
plt.ylabel("Number of News Articles")
plt.grid(axis='y', linestyle='--', alpha=0.5)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, str(height),
             ha='center', va='bottom')

plt.tight_layout()
plt.savefig("../outputs/sentiment_distribution_linkedin.png", dpi=300)
plt.show()
```
